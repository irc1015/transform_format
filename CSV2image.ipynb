```python
import numpy as np
import matplotlib.pyplot as plt

MAX_EPOCHES = 100

x = np.arange(1, MAX_EPOCHES + 1, 1)
```

```python
y_value_1 = []
y_value_2 = []

read_values_1 = np.genfromtxt('/Users/zhuzhirui/Desktop/1/1.csv', delimiter=',')#transform into numpy
read_values_2 = np.genfromtxt('/Users/zhuzhirui/Desktop/1/2.csv', delimiter=',')#transform into numpy
'''
0:train/box_loss 
1:train/obj_loss 
2:train/cls_loss 
3:metrics/precision 
4:metrics/recall
5:metrics/mAP_0.5 
6:metrics/map_0.5:0.95 
7:val/box_loss 
8:val/obj_loss 
9:val/cls_loss
10:x/lr0 
11:x/lr1 
12:x/lr2
'''

n1_1 = np.delete(read_values_1, 0, 0)#remove 1st line (name line)
n2_1 = np.delete(n1_1, 0, 1)#remove 1st column (epoch number colomn)

n1_2 = np.delete(read_values_2, 0, 0)#remove 1st line (name line)
n2_2 = np.delete(n1_2, 0, 1)#remove 1st column (epoch number colomn)

values_1 = n2_1[:MAX_EPOCHES]
values_2 = n2_2[:MAX_EPOCHES]

print('csv1 shape is {}'.format(values_1.shape))
print('csv2 shape is {}'.format(values_2.shape))
```

```python
precision_1 = [i for i in values_1[:, 3]]
precision_1 = np.stack(precision_1)

precision_2 = [i for i in values_2[:, 3]]
precision_2 = np.stack(precision_2)

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, precision_1, label='YOLOv5s')
ax.plot(x, precision_2, label='YOLOv5s6')
ax.set(xlabel='epoches', ylabel='precision')
ax.legend(loc='lower right')
plt.show()
```
